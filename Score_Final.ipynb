```python
import pandas as pd
```

```python
# 1. Charger les données
# =========================
data = pd.read_excel("hotel_reviews_preprocessed.xlsx")
```

```python
# 2. Normalisation des notes
# =========================
MAX_RATING = 5

data["score_normalized"] = data["rating"] / MAX_RATING
```

```python
# 3. Conversion du sentiment en score numérique
# =========================
sentiment_mapping = {
    "Positive": 1,
    "Negative": -1
}

data["sentiment_score"] = data["Sentiment"].map(sentiment_mapping)

# (option sécurité)
data["sentiment_score"] = data["sentiment_score"].fillna(0)
```

```python
# 4. Calcul du Score Final
# =========================
data["Score_Final"] = (
    0.5 * data["score_normalized"]
) + (
    0.5 * data["sentiment_score"] * MAX_RATING
)
```

```python
# 5. Vérification
# =========================
print(data[[
    "rating",
    "score_normalized",
    "Sentiment",
    "sentiment_score",
    "Score_Final"
]].head())
```

```python
output_path = "hotel_reviews_with_score_final.xlsx"
data.to_excel(output_path, index=False)

print("✅ Dataset avec Score Final sauvegardé")
```
